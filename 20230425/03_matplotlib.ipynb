# 데이터의 시각화
1. 라인, 바, 산점도, 히스토그램 그래프 시각화

```python
import matplotlib.pyplot as plt
```

```python
# 라인 그래프
# plt.plot([]) -> 리스트의 값이 하나만 존재하면 y축의 값으로 지정
# plt.plot([], []) -> 앞의 리스트의 값은 x축, 뒤의 리스트의 값은 y축이 된다.
plt.plot([1, 2, 3, 4])
plt.show()
```

# subplot()
1. 한 영역에 그래프를 여러 개 그려주기 위한 함수
2. subplot(행의 개수, 열의 개수, 위치)

```python
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3, 4])
plt.subplot(1, 2, 2)
plt.plot([4, 3, 2, 1])
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)
axes[0][0].plot([1, 2, 3, 4])
axes[0][1].plot([4, 3, 2, 1])
axes[1][0].plot([1, 2, 2, 1])
axes[1][1].plot([3, 4, 2, 1])
plt.show()
```

```python
x = [1, 2, 3]
y = [4, 3, 2, 1]
plt.plot(x, y)
plt.show()
# 첫번째 디맨션에 같은 양의 값이 제공되어야 그래프를 그릴 수 있음
```

```python
# x축의 이름과 y축의 이름, 그래프의 이름 지정
plt.plot([1, 2, 3, 4])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('title')
plt.show()
```

```python
# 범례 추가
plt.plot([1, 2, 3, 4], label = 'acending')
plt.plot([4, 3, 2, 1], label = 'descending')
plt.legend(loc = 'upper right')
plt.show()
```

```python
# 히스토그램 -> 데이터의 개수를 표현하는 바 그래프
data = [1, 2, 2, 2, 3, 3, 4, 4, 4, 4]
plt.hist(data)
plt.show()
```

# 문제
1. test_csv 폴더 안에 있는 오프라인.csv을 로드
2. 컬럼의 이름을 ['기준년월', '업종', '성별', '연령대', '총이용건수', '총이용금액']으로 변경
3. 기준년월을 시계열 데이터로 변경
4. index를 기준년월로 변경
5. 온라인.csv 파일도 위의 작업을 반복
6. 온라인에 업종 컬럼을 추가하여 '온라인'으로 값을 지정
7. 두 개의 데이터프레임을 유니언 결합
8. 기준년월, 업종 별 총 이용금액의 합계를 라인그래프로 시각화

```python
import pandas as pd
```

```python
offline = pd.read_csv('../test_csv/오프라인.csv', encoding='euc-kr')
```

```python
offline.columns = [
    '기준년월',
    '업종',
    '성별',
    '연령대',
    '총이용건수',
    '총이용금액'
]
```

```python
offline['기준년월'] = pd.to_datetime(offline['기준년월'], format='%Y%m')
```

```python
offline.set_index('기준년월', inplace=True)
```

```python
online = pd.read_csv('../test_csv/온라인.csv', encoding='euc-kr')
```

```python
online.columns = [
    '기준년월',
    '성별',
    '연령대',
    '총이용건수',
    '총이용금액'
]
```

```python
online['업종'] = '온라인'
```

```python
online.head()
```

```python
online['기준년월'] = pd.to_datetime(online['기준년월'], format='%Y%m')
```

```python
online.set_index('기준년월', inplace=True)
```

```python
data = pd.concat([offline, online], axis=0)
```

```python
data.sort_index(inplace=True)
```

```python
pivot_data = pd.pivot_table(
    data,
    index = data.index,
    columns= '업종',
    values= '총이용금액',
    aggfunc = 'sum'
)
```

```python
plt.rc('font', family='AppleGothic')
```

```python
pivot_data.plot()
```

```python
group_data = data.groupby([data.index, '업종']).sum()[['총이용금액']]
```

```python
group_data.unstack().plot()
```

1. 결합된 데이터프레임에서 성별을 1은 '남자', 2는 '여자'로 변경
2. 온라인 업종의 남, 여의 카드매출을 월별로 어떻게 차이가 나는지 시각화하여 표시

```python
# 성별을 1은 남자 2는 여자로 변경
data.loc[data['성별'] == 1, '성별'] = '남자'
data.loc[data['성별'] == 2, '성별'] = '여자'
data['성별'].value_counts()
```

```python
data['성별'].isna().sum()
```

```python
online_data = data.loc[data['업종'] == '온라인']
```

```python
pd.pivot_table(
    online_data,
    index = online_data.index,
    columns = '성별',
    values = '총이용금액',
    aggfunc = 'mean'
).plot()
```
