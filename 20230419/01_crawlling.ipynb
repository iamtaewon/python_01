```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# 실제 사이트에서 html 정보를 받아와서 변수에 저장
data = requests.get('https://finance.naver.com/marketindex/exchangeList.naver').text
# response [200] -> 정상
```

```python
# BeautifulSoup을 이용하여 데이터를 파싱
soup = bs(data, 'html.parser')
```

```python
print(soup)
```

```python
# find는 html안에서 해당 태그의 첫번째 정보를 출력
tbody = soup.find('tbody')
```

```python
# find_all() 해당 태그의 전체 정보를 리스트의 형태로 출력
data_list = soup.find_all('td')
```

```python
_list = []
for i in data_list:
    x = i.get_text()
    x = x.replace('\n', '')
    x = x.strip('\t')
    _list.append(x)
```

```python
sql = """
    create
    table
    `naver`(
        `통화명` varchar(16) primary key,
        `매매기준율` varchar(16),
        `현찰(구매)` varchar(16),
        `현찰(판매)` varchar(16),
        `송금(보낼때)` varchar(16),
        `송금(받을때)` varchar(16),
        `미화환산율` varchar(16)
    )
"""
db.sql_query(sql)
```
