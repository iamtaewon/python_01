```python
import pandas as pd
```

### 연습문제
1. csv 폴더에 있는 Sales Records.csv 파일 로드
2. 'Order ID', 'Total Revenue', 'Total Cost', 'Total Profit', 'Sales Channel' 컬럼을 삭제
3. 컬럼의 이름을 변경 -> ['권역', '국가', '상품종류', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']
4. 권역과 국가를 기준으로 오름차순 정렬
5. index를 초기화하고 기존의 인덱스는 삭제

```python
# 파일로드
# 상위 폴더로 이동 -> csv 폴더로 이동 -> 파일명
```

```python
sales = pd.read_csv('../csv/Sales Records.csv')
sales.head(2)
```

```python
# Order ID 컬럼 삭제
# sales = sales.drop('Order ID', axis=1) --> sales = 대신 inplac = True 함수 사용 가능 --> 변환한 데이터를 저장하는 것
# 컬럼 동시 삭제
```

```python
sales.drop(['Order ID', 'Total Revenue', 'Total Cost', 'Total Profit', 'Sales Channel'], axis = 1, inplace = True)
```

```python
# 컬럼의 이름을 변경
```

```python
sales.columns = ['권역',
                 '국가', 
                '상품종류',
                '우선순위',
                '주문일자',
                '발송일자', 
                '판매단위',
                '단가',
                '원가']
```

```python
# 권역, 국가를 기준으로 오름차순 정렬
```

```python
sales.sort_values(['권역','국가'], inplace = True)
sales
```

```python
# 인덱스를 초기화하고 기존의 인덱스를 삭제
```

```python
sales.reset_index(drop=True, inplace = True)
sales
```

- 새로운 파생 변수 생성
    1. 총단가 --> 판매단위 * 단가
    2. 총원가 --> 판매단위 * 원가
    3. 총이윤 --> 총단가 - 총원가

```python
sales['총단가'] = sales['판매단위'] * sales['단가']
```

```python
sales['총원가'] = sales['판매단위'] * sales['원가']
```

```python
sales['총이윤'] = sales['총단가'] - sales['총원가']
```

```python
sales
```

```python
# 국가 컬럼의 데이터가 항목들의 카운트를 출력하는 함수
sales['국가'].value_counts()
```

```python
# 국가라고 하는 value의 index 값만 출력
sales['국가'].value_counts().index
```

```python
# 위의 값과 같은 값
_list = sales['국가'].unique()
```

```python
# 국가 하나를 선택해서 그 국가의 총이윤의 합계
# sales.loc[sales['국가'] == 'South Korea']['총이윤']
sales.loc[sales['국가'] == 'South Korea', '총이윤'].sum()
```

```python
profit_list = {}

# 반복문 생성
for i in _list:
    # 총이윤의 합계
    total_profit = sales.loc[sales['국가'] == i, '총이윤'].sum()
    profit_list[i] = total_profit

profit_list
```

```python
# 가장 많은 총이윤을 가진 나라 순으로 정렬
pd.Series(profit_list).sort_values(ascending=False)
```

```python
profit_list.values()
```

```python
pd.DataFrame(profit_list.values(),
             index=profit_list.keys(),
             columns=['SUM'])
```

```python
# 그룹화 함수
# groupby()
# 인자 값 : 특정 컬럼을 지정
# sales.groupby('국가').sum()
# 멀티 인덱스, 멀티 컬럼
group_data = sales.groupby(['권역','국가']).sum()
```

```python
group_data.index
# 리스트 안에 튜플이 생김 --> 2차원 배열
```

### sales 데이터에서 상품의 종류에 따라 총 이윤이 가장 높은 물건은 무엇인가?
    1. sales 데이터에서 그룹화 진행 -> 그룹조건 => 상품종류
    2. 그룹화된 데이터에서 총 이윤이라는 컬럼의 데이터의 평균을 출력
    3. 총 이윤의 평균값을 기준으로 내림차순 정렬
    4. 총 이윤이 가장 높은 상품 출력

```python
# sales.groupby(['상품종류']).mean()['총이윤'] -> Series 형태
item_data = sales.groupby(['상품종류']).mean()[['총이윤']] # -> DataFrame 형태
```

```python
df = item_data.sort_values('총이윤', ascending = False)
```
